```python
import pandas as pd

data = {
    "likes": [1200, 3400, 800, 5000, 2300, 4100, 1500],
    "comments": [45, 120, 20, 300, 90, 200, 60],
    "followers": [10000]*7,
    "post_type": ["image", "video", "image", "video", "image", "video", "image"],
    "post_time": [9, 21, 14, 20, 18, 22, 11],
    "hashtags_count": [5, 10, 3, 12, 7, 11, 4]
}

df = pd.DataFrame(data)
df
```

```python
df.to_csv('../data/instagram_data.csv', index=False)
print("CSV saved successfully")
```

```python

df['engagement_rate'] = (df['likes'] + df['comments']) / df['followers']
df
```

```python
avg_engagement = df['engagement_rate'].mean()
print(f"Average Engagement Rate: {avg_engagement:.2%}")
```

data visualization

```python
import matplotlib.pyplot as plt
import seaborn as sns


sns.set(style="whitegrid")
```

```python
plt.figure(figsize=(8,5))
sns.barplot(x=df.index, y=df['engagement_rate'], hue=df['post_type'])
plt.title("Engagement Rate per Post by Post Type")
plt.xlabel("Post Index")
plt.ylabel("Engagement Rate")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x=df['hashtags_count'], y=df['engagement_rate'], hue=df['post_type'], s=100)
plt.title("Engagement Rate vs Hashtag Count")
plt.xlabel("Number of Hashtags")
plt.ylabel("Engagement Rate")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.lineplot(x=df['post_time'], y=df['engagement_rate'], marker='o')
plt.title("Engagement Rate vs Post Time")
plt.xlabel("Post Time (Hour)")
plt.ylabel("Engagement Rate")
plt.show()
```
